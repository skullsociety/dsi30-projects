## Background

The ACT (American College Testing) test is a American standardize test administered by an non-profit organisation of the same name, ACT. It is used as a quality test of college applicants required for college admission process. The applicants portfolio would need to include ACT or SAT scores along with other materials to be considered for acceptance to the university.([*source*](https://www.act.org/))

The ACT consist of four sections : English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). 

Scores for individual sections are scaled within range of 1 to 36. 1 for low score and 36 for a high score. Composite score would be calculated by the average of all section test scores, rounded to the nearest whole number. ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html))

While ACT popularity as a standardized test have been increasing since 1940, ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)) there have been a increased supporters of these test, having a view that these tests are not a good measure to determine college admittance as these tests are not adequate to determine the ability.

```python
%%time
#function for mean of an integer, input a list, titled 'mean'

# def mean(list):
#     total_number = 0
#     for number in test_list:
#         total_number += number
#     average = total_number / len(list)
#     return average

def mean(list):
    return (sum(list)/len(list))
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# function for standard deviation, input a list, titled 'std'

def std (list):
    store = 0 
    for number in list:
        store += (number - (sum(list)/len(list)))**2
    return (store/(len(list)-1))**(0.5)
```

```python
#function for removing "%" in a string, input string, output float value ie: 5% to 0.05

def nopercent (number_string_with_percent):
    return (float((int(number_string_with_percent.replace("%", "")))/100))
```

```python
#function to print max and min value for individual data in ACT 2017 data
def act_mm_test (data):
    print (f"Particiapation max values is {(data)['Participation'].max()}")
    print (f"Particiapation min values is{(data)['Participation'].min()}")
    print (f"English max values is {(data)['English'].max()}")
    if (data)['English'].max() >36:
        print (f"There is an issue here")
    print (f"English min values is {(data)['English'].min()}")
    if (data)['English'].min() <0:
        print (f"There is an issue here")
    print (f"Math max values is {(data)['Math'].max()}")
    if (data)['Math'].max() >36:
        print (f"There is an issue here")
    print (f"Math min values is {(data)['Math'].min()}")
    if (data)['Math'].min() <0:
        print (f"There is an issue here")
    print (f"Reading max values is {(data)['Reading'].max()}")
    if (data)['Reading'].max() >36:
        print (f"There is an issue here")
    print (f"Reading min values is {(data)['Reading'].min()}")
    if (data)['Reading'].min() <0:
        print (f"There is an issue here")
    print (f"Science max values is {(data)['Science'].max()}")
    if (data)['Science'].max() >36:
        print (f"There is an issue here")
    print (f"Science min values is {(data)['Science'].min()}")
    if (data)['Science'].min() <0:
        print (f"There is an issue here")
    print (f"Composite max values is {(data)['Composite'].max()}")
    t = float((data)["Composite"].max())
    tt = float((data)["Composite"].min())
    if t >36:
        print (f"There is an issue here")
    print (f"Composite min values is {(data)['Composite'].min()}")
    if tt <0:
        print (f"There is an issue here")
```

```python
#ACT mean composite scores over the years
act789cmean= {
17:act789c['composite17'].mean(),
18:act789c['composite18'].mean(),
19:act789c['composite19'].mean()
}

#ACT participation rates in US, total over the years
act789ptotal= {
17:sum(act789p['participation17']),
18:sum(act789p['participation18']),
19:sum(act789p['participation19']),
}

#ACT mean participation rates in US states, total over the years
act789pmean = {
17:act789p['participation17'].mean(),
18:act789p['participation18'].mean(),
19:act789p['participation19'].mean(),
}

def extractdatax(dictionary):
    x = []
    for k in dictionary.keys():
        x.append(k)
    return (x)

def extractdatay(dictionary):
    y = []
    for v in dictionary.values():
        y.append(v)
    return (y)


plt.figure(figsize = (12, 12))

    
plt.subplot(2,2,1)
x = extractdatax(act789cmean)
y = extractdatay(act789cmean)
ax = sns.lineplot(x,y)
ax.set_title('Mean for composite score 17 - 19')
ax.set_xlabel('years')
ax.set_ylabel('Composite Score Mean')
plt.locator_params(axis="x", integer=True, tight=True)

plt.subplot(2,2,2)
x2 = extractdatax(act789ptotal)
y2 = extractdatay(act789ptotal)
ax = sns.lineplot(x2,y2)
ax.set_title('Total participation rates 17 - 19')
ax.set_xlabel('years')
ax.set_ylabel('Participation rates')
plt.locator_params(axis="x", integer=True, tight=True)

plt.subplot(2,2,3)
x3 = extractdatax(act789pmean)
y3 = extractdatay(act789pmean)
ax = sns.lineplot(x3,y3)
ax.set_title('Average participation rates 17 - 19')
ax.set_xlabel('years')
ax.set_ylabel('Participation rates')
plt.locator_params(axis="x", integer=True, tight=True)

plt.show()
```
